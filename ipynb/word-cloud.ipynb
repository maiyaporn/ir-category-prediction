```python
from os import path
import pickle
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
cluster_c25 = '../clustering/c25/tmp/'
```

```python
topic_files = []
for n in [350]:
    fname = str(n) + '_topics'
    topic_files.append(cluster_c25 + fname)

wc = WordCloud(background_color="#EDEDED", width=800, height=600, font_path='/Users/Muk/Library/Fonts/Helvetica.dfont')
```

```python
for tf in topic_files:
    with open(tf, 'rb') as f:
        topics = pickle.load(f)
        for t, topic in enumerate(topics):
            words = [(j,i) for i, j in topic]
            wc.fit_words(words)
            plt.imshow(wc)
            #plt.axis("off")
            #plt.show()
            wc.to_file( '../clustering/c25/visualization/' + str(tf.split("/")[-1]) + '/t' + str(t) + '.png' )
```
